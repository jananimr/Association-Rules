# Import libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
# conda install -c conda-forge mlxtend
```

```python
movie1 = pd.read_csv("C:/Users/khann/OneDrive/Desktop/excel r assignments/my_movies.csv")
movie1.head(15)
```

```python
movie=movie1.iloc[:,5:15]
movie
```

# Pre-Processing
As the data is not in transaction formation 
We are using transaction Encoder

```python
movie_list=pd.get_dummies(movie)
movie_list
```

# Apriori Algorithm 

```python
frequent_itemsets = apriori(movie_list, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets = apriori(movie_list, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
rules.plot()
```

```python
rules['lift'].hist()
```

```python
rules.boxplot(column=['lift', 'confidence'])
```

```python
#####  An leverage value of 0 indicates independence. Range will be [-1 1]
### high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```
